```sql
-- KPIs GERAIS
CREATE OR REPLACE TABLE workspace.taxi_trip_gold.kpi_overall AS
SELECT
  COUNT(*) AS total_trips,
  ROUND(SUM(total_amount), 2) AS total_revenue,
  ROUND(AVG(total_amount), 2) AS avg_ticket,
  ROUND(AVG(trip_distance), 2) AS avg_distance,
  ROUND(AVG(tip_amount), 2) AS avg_tip
FROM workspace.taxi_trip_silver.silver_taxi_trips;

-- RECEITA POR DIA
CREATE OR REPLACE TABLE workspace.taxi_trip_gold.kpi_daily_revenue AS
SELECT
  DATE(pickup_datetime) AS trip_date,
  COUNT(*) AS total_trips,
  ROUND(SUM(total_amount), 2) AS total_revenue,
  ROUND(AVG(total_amount), 2) AS avg_ticket
FROM workspace.taxi_trip_silver.silver_taxi_trips
GROUP BY DATE(pickup_datetime)
ORDER BY trip_date;

-- RECEITA POR HORA
CREATE OR REPLACE TABLE workspace.taxi_trip_gold.kpi_hourly_revenue AS
SELECT
  HOUR(pickup_datetime) AS trip_hour,
  COUNT(*) AS total_trips,
  ROUND(SUM(total_amount), 2) AS total_revenue,
  ROUND(AVG(total_amount), 2) AS avg_ticket
FROM workspace.taxi_trip_silver.silver_taxi_trips
GROUP BY trip_hour
ORDER BY trip_hour;

-- COMPORTAMENTO DE PAGAMENTO
CREATE OR REPLACE TABLE workspace.taxi_trip_gold.kpi_payment_behavior AS
SELECT
  payment_type,
  COUNT(*) AS total_trips,
  ROUND(SUM(total_amount), 2) AS total_revenue,
  ROUND(AVG(total_amount), 2) AS avg_ticket,
  ROUND(AVG(tip_amount), 2) AS avg_tip
FROM workspace.taxi_trip_silver.silver_taxi_trips
GROUP BY payment_type;

-- DISTÃ‚NCIA X RECEITA
CREATE OR REPLACE TABLE workspace.taxi_trip_gold.kpi_distance_analysis AS
SELECT
  CASE
    WHEN trip_distance < 1 THEN '0-1 mile'
    WHEN trip_distance < 3 THEN '1-3 miles'
    WHEN trip_distance < 5 THEN '3-5 miles'
    ELSE '5+ miles'
  END AS distance_bucket,
  COUNT(*) AS total_trips,
  ROUND(AVG(total_amount), 2) AS avg_ticket,
  ROUND(AVG(trip_distance), 2) AS avg_distance
FROM workspace.taxi_trip_silver.silver_taxi_trips
GROUP BY distance_bucket;
```
