```sql
-- Pico de corrida x hora no dia
SELECT
  hour(pickup_datetime) AS pickup_hour,
  COUNT(*) AS total_trips
FROM workspace.taxi_trip_silver.silver_taxi_trips
GROUP BY pickup_hour
ORDER BY pickup_hour;
-- A quantidade de viagens aumenta a partir das 6h da manhã, havendo outro pico a partir das 17h. A partir das 23h, a quantidade de viagens diminui. 

-- Corridas por dia da semana - há diferenças entre dia de semana x final de semana?
SELECT
  date_format(pickup_datetime, 'E') AS day_of_week,
  COUNT(*) AS total_trips
FROM workspace.taxi_trip_silver.silver_taxi_trips
GROUP BY day_of_week
ORDER BY total_trips DESC;
-- Sexta-feira e sábado são os dias com mais viagens. Domingo é o dia com menos viagens.

-- Volume de corridas ao longo do tempo - aumento ou diminuição ao longo dos meses?
SELECT
  date_trunc('month', pickup_datetime) AS month,
  COUNT(*) AS total_trips
FROM workspace.taxi_trip_silver.silver_taxi_trips
GROUP BY month
ORDER BY month;
-- Em 2016, o volume de corridas é maior em Março. Mas devido a pouca quantidade de dados por mês, uma analise temporal de viagens (por mês ou ano) é pouco viável

-- Média de duração por período - quais horários tem corridas mais longas?
SELECT
  hour(pickup_datetime) AS pickup_hour,
  AVG(
    unix_timestamp(dropoff_datetime) - unix_timestamp(pickup_datetime)) / 60 AS avg_duration_minutes
FROM workspace.taxi_trip_silver.silver_taxi_trips
GROUP BY pickup_hour
ORDER BY pickup_hour DESC;
-- No geral, a média de viagens é bem similar entre as horas. Talvez mostre que compensa mais pegar um táxi para médias ou longas distâncias

-- Preferência de forma de pagamento
SELECT
  payment_type,
  COUNT(*) AS total_trips
FROM workspace.taxi_trip_silver.silver_taxi_trips
GROUP BY payment_type
ORDER BY total_trips DESC;
-- A maioria das viagens são pagas com cartão de crédito, seguida de pagamento em dinheiro. Logo, taxistas devem estar preparados para usar maquineta princilmente, mas também ter dinheiro trocado.

-- Valor médio pago pelos passageiros
SELECT
  AVG(total_amount) AS avg_total_amount
FROM workspace.taxi_trip_silver.silver_taxi_trips;
-- Uma média de 15$ por viagem. Se houvesse mais dados sobre outros anos, poderiamos ver a evolução dos preços ao longo do tempo.

```
