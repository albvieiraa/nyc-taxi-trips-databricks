```sql
-- Forma de pagamento x receita
SELECT
  payment_type,
  COUNT(*) AS total_trips,
  ROUND(SUM(total_amount), 2) AS total_revenue,
  ROUND(AVG(total_amount), 2) AS avg_ticket
FROM workspace.taxi_trip_silver.silver_taxi_trips
GROUP BY payment_type
ORDER BY total_revenue DESC;
-- Ticket médio é maior em pagamento por cartão de crédito (havendo gorjeta embutida)

-- Tempo de corrida x receita
SELECT
  CASE
    WHEN trip_distance < 1 THEN '0-1 mile'
    WHEN trip_distance < 3 THEN '1-3 miles'
    WHEN trip_distance < 5 THEN '3-5 miles'
    ELSE '>5 miles'
  END AS distance_bucket,
  COUNT(*) AS total_trips,
  ROUND(AVG(total_amount), 2) AS avg_total_amount
FROM workspace.taxi_trip_silver.silver_taxi_trips
GROUP BY distance_bucket
ORDER BY avg_total_amount DESC;
-- Corridas mais longas geram mais receita.

-- Horário x Lucro
SELECT
  HOUR(pickup_datetime) AS pickup_hour,
  COUNT(*) AS total_trips,
  ROUND(AVG(total_amount), 2) AS total_revenue
FROM workspace.taxi_trip_silver.silver_taxi_trips
GROUP BY pickup_hour
ORDER BY total_revenue DESC;
-- Horário de maior lucro é entre 4h e 5h da manhã; apesar de ter menos volume de corrida, são elas mais caras. Enquanto que entre 7h e 9h da manhã são as mais baratas e com volume maior.

-- Dia da semana x ticket médio
SELECT
  date_format(pickup_datetime, 'EEE') AS day_of_week,
  COUNT(*) AS total_trips,
  ROUND(AVG(total_amount), 2) AS avg_ticket
FROM workspace.taxi_trip_silver.silver_taxi_trips
GROUP BY day_of_week
ORDER BY avg_ticket DESC;
-- Domingo e Quinta-feira são os dias com ticket médio mais caro, apesar de não necessariamente representarem os dias com volume de corrida mais altos.

-- Gorgeta x tipo de pagamento
SELECT 
  payment_type,
  ROUND(AVG(tip_amount), 2) AS avg_tip_amount,
  ROUND(AVG(tip_amount / total_amount) * 100, 2) AS avg_tip_percentage
FROM workspace.taxi_trip_silver.silver_taxi_trips
WHERE tip_amount > 0
GROUP BY payment_type
ORDER BY avg_tip_percentage DESC;
-- Verificar esse último cálculo
```
